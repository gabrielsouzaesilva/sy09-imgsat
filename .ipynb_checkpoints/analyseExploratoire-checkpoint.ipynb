# Dataset analysis
- ## Descriptive Analysis:
    - **Target variable**: class
    - **Lines and columns**: (10545, 29)
    - **Variable types**: 28 quantitatives (float64) et 1 qualitative (object)
    - **Missing value analysis**: no NA

- ## Diagnostic Analysis:
    - **Visualization of the target variable**: 
        - 70% of the values are classified as Forest and there is very little data from other classes
    - **Meaning of variables**: 
        - variables that represent days appear unstandardized and with a lot of noise
        - the max_ndvi is asymmetric
    - **Variables/Target relationship**:
        - Hypothesis: the classes are centered around different ndvi values, because they are different types of vegetation/soil
        - We have noticed that on some days, some or all of the classes are centered around a point near zero or close to 8000 ndvi (See days 20140218_N and 20150109_N, for example). It is interesting to check the influence of noise.
        - The **max_ndvi** can be used to identify classes. From the definition of ndvi we see that the clouds lower the values of ndvi, then we can think, that under good atmospheric conditions, the max_ndvi represent the true value of ndvi of the region. This seems reasonable, because according to the distribution of max_ndvi by class, we see forest, grass, farm and orchard, which are vegetation with too much green, close to 8000 (high values). Additionally, water and impervious are down around 2000, which agrees well with the definition of ndvi. (See for more details on ndvi
 https://en.wikipedia.org/wiki/Normalized_difference_vegetation_index)
            - Also, we can think about creating a vegetation group with (forest, grass, farm and orchard) to study how to separate these classes

- ## Detailed analysis
    - ### Variables/Variables relationship:
        - **All/All**: 
            - There are no strongly correlated variables, but there are a few with a correlation greater than 0.5
        - **max_ndvi/tous**: 
            - We notice that the day 20140202_N is linearly strongly correlated and there are others which are less correlated, but around 0.5
            
            
- ## Null hypothesis (H0):
    - Individuals have significantly different mean of max_ndvi
    - H0 = The means of max_ndvi are equal in individuals of all classes
   

# Initial conclusions

- ## Hypothesis
    - We reject the null hypothesis except for the water and impervious classes, so these classes do not have significantly different means of max_ndvi

# Test de Student

```python
# Check class size
print(df_forest.shape)
print(df_grass.shape)
print(df_farm.shape)
print(df_orch.shape)
print(df_water.shape)
print(df_imp.shape)
```

```python
# Sampling
# Test with 53 samples (minimum of orchard) in each class
samples = 53
bal_forest = df_forest.sample(samples)
bal_grass = df_grass.sample(samples)
bal_farm = df_farm.sample(samples)
bal_water = df_water.sample(samples)
bal_imp = df_imp.sample(samples)

jeus = [bal_farm, bal_forest, bal_grass, bal_imp, bal_water, df_orch]
```

```python
# Student T-Test function
def t_test(df_a, df_b):
    alpha=0.02
    stat, p = ttest_ind(df_a, df_b)
    if p < alpha:
        return 'H0 Rejete'
    else:
        return 0
```

```python
# Student's t-test for each pair of classes
for jeu in jeus:
    for j in jeus:
        print (jeu['class'].unique(), j['class'].unique())
        print(t_test(jeu['max_ndvi'], j['max_ndvi']))
```
